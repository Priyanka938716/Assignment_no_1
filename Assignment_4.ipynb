```python
1. Write a function safe_divide(a, b) that divides two numbers and handles 
ZeroDivisionError and TypeError separately, printing an appropriate message 
for each. Use a finally block to print 'Operation attempted.' regardless of the 
outcome. Test it with valid numbers, division by zero, and a string input.
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
    except TypeError:
        print("Error: Both inputs must be numbers.")
    finally:
        print("Operation attempted.")

# Testing the function
print("Test 1: Valid numbers")
safe_divide(10, 2)

print("\nTest 2: Division by zero")
safe_divide(10, 0)

print("\nTest 3: String input")
safe_divide(10, "2")
```

```python
2. Write a function validate_age(age) that raises a custom exception InvalidAgeError 
(a class you define, inheriting from Exception) if age is negative or greater than 
120. Catch this exception in a calling block and print a user-friendly message. Test it with 
at least 3 different values.
```

```python
# Custom exception class
class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0:
        raise InvalidAgeError("Age cannot be negative.")
    if age > 120:
        raise InvalidAgeError("Age cannot be greater than 120.")
    print(f"Valid age: {age}")

# Testing with 3 different values
test_ages = [25, -5, 150]

for age in test_ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print(f"Error: {e}")
```

```python
3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated 
inputs in a list, e.g., user_inputs = ['10', 'abc', '0', '5']) to enter a number and 
divide 100 by it. Handle the following in a single try block using multiple except clauses:
a) ValueError if the input is not a number.  
b) ZeroDivisionError if the input is zero.  
c) A generic Exception as a catch-all for anything else.  

Print the result if successful, otherwise print which error occurred.
```

```python
user_inputs = ['10', 'abc', '0', '5']

for item in user_inputs:
    try:
        num = int(item)          
        result = 100 / num       
        print(f"100 / {num} = {result}")
    
    except ValueError:
        print(f"Error: '{item}' is not a number.")
    
    except ZeroDivisionError:
        print(f"Error: Cannot divide by zero. Input was '{item}'.")
    
    except Exception as e:
        print(f"Unexpected error with input '{item}': {e}")
```

```python
4. Create a 1D NumPy array of integers from 1 to 20. Perform the following:  
a) Reshape it into a 4x5 2D array.  
b) Print the sum of each row and each column.  
c) Extract all elements greater than 10 using boolean indexing.  
d) Replace all even numbers in the array with -1.
```

```python
import numpy as np

# a) Create 1D array and reshape to 4x5
arr = np.arange(1, 21)  # array from 1 to 20
arr_2d = arr.reshape(4, 5)

print("4x5 Array:\n", arr_2d)

# b) Sum of each row and each column
print("\nSum of each row:", arr_2d.sum(axis=1))
print("Sum of each column:", arr_2d.sum(axis=0))

# c) Extract elements greater than 10 using boolean indexing
greater_than_10 = arr_2d[arr_2d > 10]
print("\nElements greater than 10:", greater_than_10)

# d) Replace all even numbers with -1
arr_2d[arr_2d % 2 == 0] = -1
print("\nArray after replacing even numbers with -1:\n", arr_2d)
```

```python
5. Create two NumPy arrays A and B as shown below and perform:  
`A = np.array([[1, 2], [3, 4]])`  
`B = np.array([[5, 6], [7, 8]])`  

a) Element-wise addition, subtraction, and multiplication.  
b) Matrix multiplication using `np.dot()` or the `@` operator.  
c) Transpose of A.  
d) The determinant and inverse of A using `np.linalg`.
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# a) Element-wise operations
print("Element-wise Addition:\n", A + B)
print("\nElement-wise Subtraction:\n", A - B)
print("\nElement-wise Multiplication:\n", A * B)

# b) Matrix multiplication
print("\nMatrix Multiplication using @:\n", A @ B)
print("\nMatrix Multiplication using np.dot():\n", np.dot(A, B))

# c) Transpose of A
print("\nTranspose of A:\n", A.T)

# d) Determinant and inverse of A
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)

print(f"\nDeterminant of A: {det_A:.2f}")
print("\nInverse of A:\n", inv_A)
```

```python
6. Generate a NumPy array of 15 random integers between 1 and 100 using `np.random.randint()` (set a seed of 42 for reproducibility). Then:  
a) Find the mean, median, standard deviation, min, and max.  
b) Sort the array in descending order.  
c) Find the indices of the 3 largest values using `np.argsort()`.  
d) Normalize the array so all values lie between 0 and 1.
```

```python
import numpy as np

np.random.seed(42)
arr = np.random.randint(1, 101, size=15)

print("Original Array:\n", arr)

print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))

desc_sorted = np.sort(arr)[::-1]
print("\nSorted in Descending Order:\n", desc_sorted)

indices_3_largest = np.argsort(arr)[-3:][::-1]
print("\nIndices of 3 largest values:", indices_3_largest)
print("3 largest values:", arr[indices_3_largest])

normalized = (arr - arr.min()) / (arr.max() - arr.min())
print("\nNormalized Array (0 to 1):\n", normalized)
```

```python
7. Create a Pandas DataFrame from the dictionary below and perform the following:
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], 
         "Age": [23, 21, 25, 22, 24], 
         "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"], 
         "Score": [88, 92, 79, 95, 67], }
a) Display the first 3 rows and the data types of each column.  
b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').  
c) Sort the DataFrame by Score in descending order.  
d) Filter rows where Department is 'AI'.
```

```python
import pandas as pd

data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], 
         "Age": [23, 21, 25, 22, 24], 
         "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"], 
         "Score": [88, 92, 79, 95, 67], }

df = pd.DataFrame(data)

print("First 3 rows:\n", df.head(3))
print("\nData types:\n", df.dtypes)

df['Grade'] = df['Score'].apply(lambda x: 'A' if x >= 90 else 'B' if x >= 75 else 'C')
print("\nDataFrame with Grade column:\n", df)

df_sorted = df.sort_values(by='Score', ascending=False)
print("\nSorted by Score (descending):\n", df_sorted)

df_ai = df[df['Department'] == 'AI']
print("\nRows where Department is 'AI':\n", df_ai)
```

```python
8. Using the same DataFrame from Q7:
a) Group by Department and find the average Score for each.  
b) Find the row with the highest Score using `idxmax()`.  
c) Rename the column Score to Marks.  
d) Drop the Age column and display the resulting DataFrame.
```

```python
import pandas as pd

# DataFrame from Q7
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], 
         "Age": [23, 21, 25, 22, 24], 
         "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"], 
         "Score": [88, 92, 79, 95, 67], }

df = pd.DataFrame(data)

# a) Group by Department and find average Score
avg_score = df.groupby('Department')['Score'].mean()
print("Average Score by Department:\n", avg_score)

# b) Row with highest Score using idxmax()
idx_max = df['Score'].idxmax()
highest_row = df.loc[idx_max]
print("\nRow with highest Score:\n", highest_row)

# c) Rename Score to Marks
df = df.rename(columns={'Score': 'Marks'})
print("\nDataFrame after renaming Score to Marks:\n", df.head())

# d) Drop Age column
df = df.drop(columns=['Age'])
print("\nFinal DataFrame after dropping Age:\n", df)
```

```python
9. Create a DataFrame with some missing values as shown below and handle them:*
import numpy as np
data = { "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
         "Price": [10, np.nan, 5, 8, np.nan],
         "Stock": [100, 50, np.nan, 30, 20], }
a) Check the number of missing values in each column using `isnull().sum()`.  
b) Fill missing Price values with the column mean.  
c) Drop rows where Stock is still missing.  
d) Save the cleaned DataFrame to a CSV file named `cleaned_products.csv` without the index.
```

```python
import pandas as pd
import numpy as np

data = { "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
         "Price": [10, np.nan, 5, 8, np.nan],
         "Stock": [100, 50, np.nan, 30, 20], }

df = pd.DataFrame(data)

# a) Check missing values in each column
print("Missing values per column:\n", df.isnull().sum())

# b) Fill missing Price values with column mean
price_mean = df['Price'].mean()
df['Price'] = df['Price'].fillna(price_mean)
print("\nAfter filling Price with mean:\n", df)

# c) Drop rows where Stock is still missing
df = df.dropna(subset=['Stock'])
print("\nAfter dropping rows with missing Stock:\n", df)

# d) Save to CSV without index
df.to_csv('cleaned_products.csv', index=False)
print("\nSaved to cleaned_products.csv")
```
